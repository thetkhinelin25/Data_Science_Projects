### Spliting the data into Independant and Dependant variables

```python
X = df_data_1.iloc[:, :-1].values
y = df_data_1.iloc[:, -1].values
```

### Spliting Train_set and Test_set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

### Creating a Linea Regression model

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()

# Train the model
regressor.fit(X_train, y_train)
```

### Predicting the result of X_test and X_train

```python
# Predict the result of X_test
y_predict_for_X_test = regressor.predict(X_test)

# Predict the result of X_train
y_predict_for_X_train = regressor.predict(X_train)
```

### Plotting Salaries Vs Experience graph based on y_predict_for_X_train

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, y_predict_for_X_train, color = 'blue')
plt.title('Salaries Vs Experiences')
plt.xlabel('Experiences')
plt.ylabel('Salaries')
plt.ylim(0)
plt.show()
```

### Plotting Salaries Vs Experience graph based on y_predict_for_X_test

```python
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, y_predict_for_X_train, color = 'blue')
plt.title('Salaries Vs Experiences')
plt.xlabel('Experiences')
plt.ylabel('Salaries')
plt.ylim(0)
plt.show()
```
